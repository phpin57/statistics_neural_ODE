```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def func(t, y, c):
    return -c * np.exp(y) * (np.exp(y) + 1)

def compare(x,c):
    return 1-np.log(np.exp(1)*c*x+1)

max_v=100

t_span = [0, max_v]
f_init = [1] 

c = 2

sol = solve_ivp(func, t_span, f_init, args=(c,))

t = sol.t
f = sol.y[0]

x = np.linspace(0, max_v,100)
y_compare = compare(x, c)
#plt.plot(x, y_f, label=f'f(x) avec c={c}')
plt.plot(x, y_compare, label=f'compare(x,c) avec c={c}')

plt.plot(t, f)
plt.xlabel('s')
plt.ylabel('a_i')
plt.title('Graphe de a_i en fonction du temps')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

c = 1.0

# Fonction qui dfinit les quations diffrentielles couples
def equations(x, y):
    f, g = y
    df_dx = -c * np.exp(f) * (np.exp(f) + np.cos(g))
    dg_dx = c * np.exp(f) * np.sin(g)
    return [df_dx, dg_dx]

# Conditions initiales
initial_conditions = [1.0, 1.0]

# Intervalle de temps sur lequel rsoudre les quations diffrentielles
x_span = [0, 10]

# Rsolution des quations diffrentielles
solution = solve_ivp(equations, x_span, initial_conditions, t_eval=np.linspace(x_span[0], x_span[1], 1000))

# Extraction des solutions pour f(x) et g(x)
f = solution.y[0]
g = solution.y[1]

# Trac des courbes de f(x) et g(x)
plt.plot(solution.t, f, label='f(x)')
plt.plot(solution.t, g, label='g(x)')
plt.axhline(y=np.pi, color='indianred', linestyle='--', label=r'$y=\pi$')
plt.axhline(y=0, color='salmon', linestyle='--', label=r'$y=0$')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x), g(x)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
from sklearn.metrics import r2_score

# Function that defines the coupled differential equations
def equations(x, y, d):
    f, g = y
    df_dx = -c * np.exp(f) * (np.exp(f) - d*np.cos(g))
    dg_dx = -c * np.exp(f) * np.sin(g)*d
    return [df_dx, dg_dx]

# Initial conditions
initial_conditions = [1.0, 1.0]

# Time interval on which to solve the differential equations
x_span = [0, 10]

# Values of d for which to solve the differential equations
d_values = np.linspace(-5, 5, 100)

# Calculate the regression coefficients and R-squared values for each value of d
slope_values = []
intercept_values = []
r2_values = []
for d in d_values:
    # Solve the differential equations for the current value of d
    solution = solve_ivp(equations, x_span, initial_conditions, args=(d,), t_eval=np.linspace(x_span[0], x_span[1], 1000))
    f = solution.y[0]
    log_diff = -np.log(np.abs(f - 0))
    
    # Calculate the regression coefficients and R-squared value
    slope, intercept = np.polyfit(solution.t, log_diff, 1)
    log_diff_pred = slope * np.array(solution.t) + intercept
    r2 = r2_score(log_diff, log_diff_pred)
    
    # Append the values to the lists
    slope_values.append(slope)
    intercept_values.append(intercept)
    r2_values.append(r2)

# Plot the R-squared values as a function of d
plt.plot(d_values, r2_values)
plt.xlabel('d')
plt.ylabel('R-squared')
plt.show()
```

```python
slopes = []

# Loop over range of d values
for d in d_values:
    # Solve differential equations
    solution = solve_ivp(equations, x_span, initial_conditions, t_eval=np.linspace(x_span[0], x_span[1], 1000), args=(d,))
    f = solution.y[0]
    log_diff = -np.log(np.abs(f - 0))

    # Calculate linear regression
    slope, _ = np.polyfit(solution.t, log_diff, 1)
    slopes.append(slope)

# Plot slopes as a function of d
plt.plot(d_values, slopes)
plt.xlabel('d')
plt.ylabel('Slope')
plt.show()
```

## Convergence dans le cas gnral en doublant la dimension

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Fonction qui dfinit les quations diffrentielles couples avec le paramtre d
def equations(x, y, d):
    f, g = y
    df_dx = -c * np.exp(f) * (np.exp(f) - d*np.cos(g))
    dg_dx = -c * np.exp(f) * np.sin(g)*d
    return [df_dx, dg_dx]

# Conditions initiales
initial_conditions = [1.0, 1.0]

# Intervalle de temps sur lequel rsoudre les quations diffrentielles
x_span = [0, 1000]

# Valeurs de d pour lesquelles on veut tracer les courbes
d_values = np.linspace(-0.3, 0.3, 11)

target_0 = 0
target_pi = np.pi

for d in d_values:
    # Rsolution des quations diffrentielles avec la valeur courante de d
    solution = solve_ivp(lambda x, y: equations(x, y, d), x_span, initial_conditions, t_eval=np.linspace(x_span[0], x_span[1], 1000))
    # Extraction des solutions pour f(x) et g(x)
    f = solution.y[0]
    g = solution.y[1]
    # Calcul de la distance  la cible pour la dernire valeur de g
    distance_0 = np.abs(g[-1] - target_0)
    distance_pi = np.abs(g[-1] - target_pi)
    
    # Affichage de l'cart entre la dernire valeur de g et la cible la plus proche
    if distance_0 < distance_pi:
        print(f'Pour d={d}, la dernire valeur de g est plus proche de 0, l\'cart est de {distance_0:.2f}')
    else:
        print(f'Pour d={d}, la dernire valeur de g est plus proche de pi, l\'cart est de {distance_pi:.2f}')
    print(f'exp(f({solution.t[-1]:.2f})) = {np.exp(f[-1]):.2f}')
    plt.plot(solution.t, f, label=f'f(x), d={d}')
    plt.plot(solution.t, g, label=f'g(x), d={d}')


# Trac des lignes horizontales
plt.axhline(y=np.pi, color='indianred', linestyle='--', label=r'$y=\pi$')
plt.axhline(y=0, color='salmon', linestyle='--', label=r'$y=0$')

# Affichage de la lgende et des tiquettes d'axe
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x), g(x)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from tqdm import tqdm

def equations(x, y, d):
    f, g = y
    df_dx = -c * np.exp(f) * (np.exp(f) - d*np.cos(g))
    dg_dx = -c * np.exp(f) * np.sin(g)*d
    return [df_dx, dg_dx]

# Dfinition de la fonction qui calcule les valeurs de g pour un d et un x donn
def f_g_value(d, x):
    # Rsolution des quations diffrentielles avec la valeur courante de d
    solution = solve_ivp(lambda x, y: equations(x, y, d), x_span, initial_conditions, t_eval=[x])
    # Extraction de la solution pour g(x)
    f=solution.y[0][0]
    g = solution.y[1][0]
    return f,g

c = 0.1

# on initialise avec l'identit relle
initial_conditions = [1.0, 1.0]

x_span = [0, 1000]

# Valeurs de d et de x pour lesquelles on veut tracer la heatmap
d_values = np.linspace(-5, 5, 101)
x_values = np.linspace(x_span[0], x_span[1], 100)

# Calcul des valeurs de g pour chaque couple (d, x)
g_matrix = np.zeros((len(d_values), len(x_values)))
f_matrix = np.zeros((len(d_values), len(x_values)))


for i, d in enumerate(tqdm(d_values)):
    for j, x in enumerate(x_values):
        f_matrix[i,j],g_matrix[i, j] = f_g_value(d, x)
```

```python
# Affichage de la heatmap
plt.imshow(g_matrix, extent=[x_span[0], x_span[1], d_values[0], d_values[-1]], cmap='RdBu', aspect='auto', origin='lower')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('d')
plt.title('Heatmap de v (partie imaginaire)')
plt.show()
```

```python
# Affichage de la heatmap
plt.imshow(f_matrix.T, origin='lower', aspect='auto', cmap='coolwarm', extent=[d_values[0], d_values[-1], x_values[0], x_values[-1]])
plt.colorbar(label='u(x)')
plt.xlabel('d')
plt.ylabel('x')
plt.title('Evolution de u (partie relle) ')
plt.show()
```

```python
exp_f_values = np.exp(f_matrix)

plt.imshow(exp_f_values, cmap='coolwarm', aspect='auto', extent=[x_span[0], x_span[1], d_values[0], d_values[-1]], origin='lower')
plt.colorbar()

# Ajout de la lgende et des tiquettes d'axe
plt.legend()
plt.xlabel('x')
plt.ylabel('d')
plt.title('Evolution de exp(u) en fonction de x et d')

plt.show()
```

```python
# Create a range of x values
x = np.linspace(0, 3, 1000)

# Define the initial conditions for f and g
initial_conditions = [1.0, 1.0]

# Define the values of d to plot
d_values = np.linspace(-5, 5, 21)

# Define a colormap to map each value of d to a color
cmap = plt.get_cmap('jet')

# Loop over each value of d and plot a curve with a color based on the value of d
for i, d in enumerate(d_values):
    # Define a function to pass to the solver that includes the value of d
    equations_with_d = lambda x, y: equations(x, y, d)
    # Solve the equations with the current value of d
    solution = solve_ivp(equations_with_d, [0, 10], initial_conditions, t_eval=x)
    # Extract the solutions for f and g
    f = solution.y[0]
    g = solution.y[1]
    # Plot the solution for f, with a color based on the value of d
    plt.plot(x, np.exp(f), color=cmap(i/len(d_values)), label=f'd = {d:.2f}')

# Add a legend and axis labels
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.xlabel('x')
plt.ylabel(r'$exp(u_d(s))$')
plt.title('Evolution of exp(u) (real part) for different values of d')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp

# Paramtres constants
c = 1

# Fonctions pour les quations diffrentielles
def equations(x, y, d):
    f, g = y
    df_dx = -c * np.exp(f) * (np.exp(f) - d*np.cos(g))
    dg_dx = -c * np.exp(f) * np.sin(g)*d
    return [df_dx, dg_dx]

# Conditions initiales
initial_conditions = [1.0, 1.0]

# Intervalle de temps sur lequel rsoudre les quations diffrentielles
x_span = [0, 10]

# Valeurs de d pour lesquelles on veut tracer les courbes
d_values = np.linspace(-5, 5, 11)

# Couleurs pour chaque courbe
colors = plt.cm.jet(np.linspace(0, 1, len(d_values)))

# Trac de la courbe pour chaque valeur de d
for i, d in enumerate(d_values):
    # Rsolution des quations diffrentielles avec la valeur courante de d
    solution = solve_ivp(lambda x, y: equations(x, y, d), x_span, initial_conditions, t_eval=np.linspace(x_span[0], x_span[1], 100))
    # Extraction des solutions pour f(x) et g(x)
    g = solution.y[1]
    
    # Trac de la courbe avec la couleur correspondante
    plt.plot(solution.t, g, label=f'd={round(d,3)}', color=colors[i])
    
# Affichage de la lgende et des tiquettes d'axe
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.xlabel('s')
plt.ylabel(r'$v_d(s)$')
plt.title('Evolution of v (imaginary part) for different values of d')
plt.show()
```
